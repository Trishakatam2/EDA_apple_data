```python
#importing libraries
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#loading dataset
my_data=pd.read_csv("apple_products.csv")
```

```python
#reading first five rows
my_data.head()
```

```python
#reading last five rows
my_data.tail()
```

```python
#checking for null values
my_data.isnull()
```

```python
my_data.isnull().sum()
```

```python
#heatplot
sns.heatmap(my_data.isnull())
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=my_data["Product Name"] , y=my_data["Sale Price"])
```

```python
sns.scatterplot(x=my_data["Sale Price"],y=my_data["Mrp"])
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x=my_data["Product Name"],y=my_data["Star Rating"])
```

```python
sns.distplot(a=my_data["Discount Percentage"],kde=False)
```

```python
sns.kdeplot(data=my_data["Star Rating"], shade=True)
```

```python
sns.jointplot(x=my_data["Sale Price"],y=my_data["Discount Percentage"], kind="kde")
```

```python
sns.lineplot(x=my_data["Sale Price"],y=my_data["Mrp"],hue=my_data["Discount Percentage"])
```

```python
sns.pairplot(my_data,kind="scatter")
```

```python
sns.regplot(x=my_data["Sale Price"],y=my_data["Star Rating"])
```

```python
sns.lmplot(data=my_data,x="Sale Price",y="Mrp",hue="Star Rating")
```
